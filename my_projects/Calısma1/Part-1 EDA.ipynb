# 36 Fuel 

     Fuel sütunu temizlendi ve yakıt tipi bilgisi; Benzine, Diesel, Others, CNG/LPG, Electric olarak kategorize edildi.

```python
df[['Fuel']].sample(10)
```

```python
df['Fuel'].value_counts()
```

```python
df.Fuel=df.Fuel.str[1].str.split('/').str[0]
```

```python
from collections import Counter

fuel_type=[]
for i in df.Fuel:
    fuel_type.append(i) 

Counter(fuel_type)
```

```python
Diesel=['Diesel','Diesel (Particulate Filter)','Biodiesel']
Benzine=['Gasoline','Super 95','Regular','Benzine E10 91 ','Super Plus 98','Regular/Benzine 91','Super E10 95','Super Plus E10 98','Super 95 (Particulate Filter)','Gasoline (Particulate Filter)','Super Plus E10 98 (Particulate Filter)','Regular/Benzine 91 (Particulate Filter)','Super E10 95 (Particulate Filter)','Regular/Benzine E10 91 (Particulate Filter)','Super Plus 98 (Particulate Filter)']
CNG_LPG=['CNG','CNG (Particulate Filter)','LPG','Liquid petroleum gas (LPG)','Domestic gas H','Biogas']
Electric=['Electric','Electric/Gasoline']


def fuel_type (a):
            
    if a in Diesel :
            return 'Diesel'
    elif a in Benzine :
            return 'Benzine'
    elif a in CNG_LPG :
        return 'CNG/LPG'
    elif a in Electric:
        return 'Electric'
    else:
        return 'Others'
```

```python
df.Fuel=df.Fuel.apply(fuel_type)
```

```python
df.Fuel.value_counts()
```
